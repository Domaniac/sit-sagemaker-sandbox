```python
from transformers import pipeline
from PIL import Image

# Load the pre-trained object detection model
object_detector = pipeline("object-detection")
print("Model loaded successfully.")
```

```python
# Helper function to print results
def analyze_image(image_path):
    print(f"--- Analyzing {image_path} ---")
    try:
        img = Image.open(image_path)
        results = object_detector(img)

        if not results:
            print("No objects detected.")
            return

        for i, item in enumerate(results):
            print(f"  Object {i+1}: {item['label']}")
            print(f"  Confidence: {item['score']:.4f}")
            print(f"  Box: {item['box']}")

    except FileNotFoundError:
        print(f"Error: Image file not found at {image_path}. Please add it to the 'data' folder.")
```

```python
# Analyze the first image
analyze_image('data/image1.jpg')
```

```python
# Analyze the second image
analyze_image('data/image2.jpg')
```
